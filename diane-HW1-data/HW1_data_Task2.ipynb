# 4. Plot a simple diagram to display closing prices for GOOG stock.  

```python
#to make a diagram, I want to use matplotlib and use alias name plt
import matplotlib.pyplot as plt
```

```python
#to see plot of the whole data in a diagram
data.plot()
```

```python
#to view particular closing prices from the data and it's corresponding date period
data['close']
```

```python
#to present in a diagram the data above 
from matplotlib import style
%matplotlib notebook

#to name a variable calling the data specifically the values under close column
msft_hx = data['close']

#select the plot style
style.use('ggplot')

#plot the data "msft_hx" and set the line width to size 1 in blue color
plt.plot(msft_hx,'b',linewidth=1)

#to set x axis "Date"
plt.xlabel('Date')

#to set y axis to "Closing Prices"
plt.ylabel('Closing Prices')

#to specify the title of the graph
plt.title('Microsoft Historical Stock \n 2020.01.01 - 2021.01.01')

#to view the graph
plt.show
```

# --The diagram clearly shows that, between March 2020 and April 2020, the closing prices for Microsoft Stock decreased significantly. This coincided with the start of the Global Pandemic.
